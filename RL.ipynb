```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
X, y = make_regression(n_samples=10000, n_features=1, noise=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2 , random_state=10)
```

```python
plt.scatter(X_train ,y_train)
```

: Implémentation des Moindres Carrés Ordinaires (MCO)

```python
class LinearRegression:
    def __init__(self):
        pass
        
    def fit(self, X, y):
        n, p = X.shape
        vecteur = np.ones(n).reshape(-1, 1)
        X = np.concatenate([vecteur, X], axis=1)
        self.beta = ((inv(X.T @ X) @ X.T) @ y.reshape(-1, 1)).flatten()
        #print(self.beta)
        
    def _predict(self, new_obs):
        new_obs = np.insert(new_obs, 0, 1)
        return (new_obs @ self.beta).item()
        
    def predict(self, X):
        return np.array([self._predict(x) for x in X])

    def score(self, X, y):
        y_predict = self.predict(X)
        mse = np.mean((y - y_predict)**2)
        return mse.item()

```

```python
model = LinearRegression()
model.fit(X_train, y_train)# entrain et trouve les paramettre
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_train)
plt.scatter(X_train ,y_train)
plt.plot(X_train, y_pred, c="r")
```
